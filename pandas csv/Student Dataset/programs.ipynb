program 8= Update Student Marks

```python


import pandas as pd
df = pd.read_csv("students.csv")

student_id = int(input("Enter Student ID: "))
new_marks = int(input("Enter New Marks: "))

df.loc[df["ID"] == student_id, "Marks"] = new_marks

print(df)
```
