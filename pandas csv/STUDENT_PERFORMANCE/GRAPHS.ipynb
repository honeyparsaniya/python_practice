1.bar chart


```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("Students.csv")

avg=df.groupby("gender")["math score"].mean()

plt.bar(avg.index,avg.values)

plt.title("average math score")
plt.xlabel("Gender")
plt.ylabel("Math Score")
```

2.pie chart
-lunch type persentage

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("students.csv")

lunch = df["lunch"].value_counts()

plt.pie(
    lunch.values,
    labels=lunch.index,
    autopct="%1.1f%%"
)

plt.title("Lunch Type Percentage")

plt.show()
```

3.histogram
-math score distribution

```python
import matplotlib.pyplot as plt

plt.hist(df["math score"])


plt.title("Math score distribution")
plt.xlabel("math score")
plt.ylabel("Student")
```

4.scatter plot
-math vs reading score

```python

import matplotlib.pyplot as plt

plt.scatter(
    df["math score"],
    df["reading score"]
)

plt.title("Math vs Reading Score")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")

plt.show()
```
