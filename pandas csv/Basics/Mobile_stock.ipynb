4. Mobile Stock Management

```python
import pandas as pd

mobile = []
stock = []

n = int(input("Enter number of mobiles: "))

for i in range(n):
    m = input("Enter mobile name: ")
    s = int(input("Enter stock quantity: "))

    mobile.append(m)
    stock.append(s)

df = pd.DataFrame({
    "Mobile": mobile,
    "Stock": stock
})

print("\nMobile Stock")
print(df)

print("\nLow Stock Mobiles")
print(df[df["Stock"] < 5])
```
