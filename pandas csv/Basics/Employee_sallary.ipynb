2. Employee Salary System

```python
import pandas as pd

employee = []
salary = []

n = int(input("Enter number of employees: "))

for i in range(n):
    e = input("Enter employee name: ")
    s = int(input("Enter salary: "))

    employee.append(e)
    salary.append(s)

df = pd.DataFrame({
    "Employee": employee,
    "Salary": salary
})

print("\nEmployee Details")
print(df)

print("\nTotal Salary =", df["Salary"].sum())

print("\nHighest Salary")
print(df[df["Salary"] == df["Salary"].max()])
```
