1. Student Marks Analysis

```python
import pandas as pd

name = []
marks = []

n = int(input("Enter number of students: "))

for i in range(n):
    n1 = input("Enter student name: ")
    m1 = int(input("Enter marks: "))

    name.append(n1)
    marks.append(m1)

df = pd.DataFrame({
    "Name": name,
    "Marks": marks
})

print("\nStudent Data")
print(df)

print("\nAverage Marks =", df["Marks"].mean())

print("\nTopper Student")
print(df[df["Marks"] == df["Marks"].max()])
```
