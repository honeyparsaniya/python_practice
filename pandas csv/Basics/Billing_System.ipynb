3. Product Billing System

```python
import pandas as pd

product = []
price = []
qty = []

n = int(input("Enter number of products: "))

for i in range(n):
    p = input("Enter product name: ")
    pr = int(input("Enter price: "))
    q = int(input("Enter quantity: "))

    product.append(p)
    price.append(pr)
    qty.append(q)

df = pd.DataFrame({
    "Product": product,
    "Price": price,
    "Quantity": qty
})

df["Total"] = df["Price"] * df["Quantity"]

print("\nBill Details")
print(df)

print("\nGrand Total =", df["Total"].sum())
```
