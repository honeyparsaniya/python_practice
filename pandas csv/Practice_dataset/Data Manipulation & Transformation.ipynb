```python
import pandas as pd
df=pd.read_csv("train.csv")

print("\noriginal column:\n")
print(df.head())

new_df=df.drop('Cabin',axis=1)

print("\nData after removing duplicate column:\n")

print(new_df.head())
```

2.apply specific function 

```python
import pandas as pd


df = pd.read_csv("train.csv")

def add_tax(fare):

    return fare + (fare * 0.10)

df['New_Fare'] = df['Fare'].apply(add_tax)

print("\nUpdated Fare:\n")

print(df[['Name', 'Fare', 'New_Fare']].head())
```
