GRAPHS

1.line graph

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("train.csv")

df["Age"].fillna(df["Age"].mean(),inplace=True)

plt.plot(df["Age"])

plt.title("passenger age trend")
plt.xlabel("passanger")
plt.ylabel("Age")
```

2.bar chart
-count male and female

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")

gender = df["Sex"].value_counts()

plt.bar(gender.index, gender.values)

plt.title("Gender Count")
plt.xlabel("Gender")
plt.ylabel("Count")
```

3.pichart
-Survival Percentage

```python
import matplotlib.pyplot as plt
survival = df["Survived"].value_counts()

plt.pie(
    survival.values,
    labels=["Dead","Alive"],
    autopct="%1.1f%%"
)
plt.title("survival persentage")

plt.show()
```

4.Histogram
-Age distribution

```python
import matplotlib.pyplot as plt

df["Age"].fillna(df["Age"].mean(), inplace=True)

plt.hist(df["Age"])

plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Count")
```

5.Scatter plot
-age vs fare

```python
import matplotlib.pyplot as plt

df["Age"].fillna(df["Age"].mean(),inplace=True)

plt.scatter(df["Age"],df["Fare"])

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
```

6.box plot
-Fare Spread

```python
import matplotlib.pyplot as plt

plt.boxplot(df["Fare"])

plt.title("Fare Box Plot")

plt.show()
```

7.heat map

```python

import seaborn as sns
import matplotlib.pyplot as plt



numeric_df = df.select_dtypes(include=["number"])

sns.heatmap(numeric_df.corr(), annot=True)

plt.title("Heatmap Correlation")

plt.show()
```
