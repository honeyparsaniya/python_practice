1.find null value in dataset(using isnull)

```python
import pandas as pd
#value avilable=false
#value does not avilable=true

df=pd.read_csv("train.csv")

print("\n===checking null value===\n")

print(df.isnull())
```

2.find null value using (not null)

```python
import pandas as pd

print("\n===checking null value===\n")

print(df.notnull())
```

3.fill value at empty space(using fillna)

```python
import pandas as pd

value=input(input("enter value to fill null data="))
new_df=df.fillna(value)

print("\n===data after fill null value===\n")
print(new_df)
```

4.remove duplicate data(using drop_duplicates)

```python
import pandas as pd

df=pd.read_csv("train.csv")

print("\n====remove duplicate value===\n")



new_df = df.drop_duplicates()

print("\n===After Removing Duplicates:===\n")
print(new_df)
```

5.change column name

```python
import pandas as pd

print("\n===original column===\n")
print(df.columns)

old_name=input("enter old column name=")
new_name=input("enter new name=")

new_df=df.rename(columns={old_name:new_name})

print("\n==updated column==\n")
print(new_df.columns)
```

6.change specific value(using replace)

```python
import pandas as pd



old_value=input("enter old value=")
new_value=input("enter new value=")

new_df=df.replace(old_value,new_value)

print("\n===updated data===\n")
print(new_df)
```
