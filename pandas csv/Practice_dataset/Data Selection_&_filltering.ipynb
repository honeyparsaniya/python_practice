1.select any single column

```python
import pandas as pd

df=pd.read_csv("train.csv")

c_name=input("enter column name=")

print("\nselected column=\n")

print(df[c_name])
```

2.print multiple column

```python
import pandas as pd

print("\n===print 3 column from data set===\n")
c_name1=input("enter first column name=")
c_name2=input("enter second column name=")
c_name3=input("enter third column name=")

print("\n---selected column---\n")

print(df[[c_name1,c_name2,c_name3]])

```

3.print selected row using loc

```python
import pandas as pd

row=int(input("enter row number to print="))

print("\n---selected row using loc===\n")

print(df.loc[row])
```

4.check value inside data or not using(df.isin)

```python
import pandas as pd
print("\n===available ticket id in data set===\n")
print(df['Ticket'])

t_id=input("enter ticket id to check=")

filter_data=df[df['Ticket'].isin(['t_id'])]

print("\n===filter data===\n")
print(filter_data)
```

5.print selected range using between

```python
import pandas as pd

start=int(input("enter start age="))
end=int(input("enter end age="))

f_data=df[df['Age'].between(start,end)]

print("\n===filter data===\n")
print(f_data)
```
